```python
import matplotlib.pyplot as plt
```

```python
# histogram chart for math score
plt.hist(df['math score'], bins=30)
plt.xlabel('math score')
plt.ylabel('Frequency')
plt.title('math score Distribution')
plt.show()
```

```python
# histogram chart for reading score
plt.hist(df['reading score'], bins=30)
plt.xlabel('reading score')
plt.ylabel('Frequency')
plt.title('reading score Distribution')
plt.show()
```

```python
# histogram chart for writing score
plt.hist(df['writing score'], bins=30)
plt.xlabel('writing score')
plt.ylabel('Frequency')
plt.title('writing score Distribution')
plt.show()
```

```python
average_score=df.groupby('gender')[['math score', 'reading score', 'writing score']].mean() 
```

```python
average_score
```

```python
# bar chart of average score by gender
average_score = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
average_score.plot(kind='bar', figsize=(8,4))
plt.xlabel('gender')
plt.ylabel(' Score')
plt.title('average Score by Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Based on the average score of test preparation course of none and completed, it shows that there is improvement in the performance of those that completed the test preparation course.
```
